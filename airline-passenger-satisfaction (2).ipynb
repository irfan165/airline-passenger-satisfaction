```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
from matplotlib import cm
target_colors = ['#800000','#248594']
```

```python
df=pd.read_csv('/kaggle/input/airline-passenger-satisfaction/train.csv')
df.head()
```

```python
df.info()
```

```python
fig=plt.figure(figsize=(15,5))
plt.suptitle('Target distribution', fontsize='14',fontname='moonspace')
grid=gridspec.GridSpec(nrows=1,ncols=2,figure=fig)
ax1 = fig.add_subplot(grid[0,:1])
ax1.set_title('Satisfaction Count')
sns.countplot(x='satisfaction', data=df, ax=ax1, hue='satisfaction',palette =target_colors)

ax2 = fig.add_subplot(grid[0,1:])
ax2.set_title('Target distribution')
label = list(df['satisfaction'].value_counts().index)
value = list(df['satisfaction'].value_counts().values)
ax2.pie(value,labels=label,autopct='%1.1f%%',explode=(0,0.2),startangle=90,colors=target_colors)
ax2.axis('equal')

plt.show()
```
